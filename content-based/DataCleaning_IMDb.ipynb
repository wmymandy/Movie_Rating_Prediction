```python
import pandas as pd
df = pd.read_csv('imdb.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
# clean up all the NaN and \N
df = df.dropna()

clean_list = ['startYear','runtimeMinutes','genres','directors','writers']
for cols_name in clean_list:
    # Get names of indexes for which column startYear has value \N
    indexNames = df[ df[cols_name] == '\\N' ].index
    # Delete these row indexes from dataFrame
    df.drop(indexNames , inplace=True)
```

```python
df.shape
```

```python
df
```

```python
# drop all the rows that total votes <100
# Get names of indexes for which column numVotes has value <100
df['numVotes'] = df['numVotes'].astype('int')
indexNames = df[ df['numVotes'] < 100 ].index
# Delete these row indexes from dataFrame
df.drop(indexNames , inplace=True)
```

```python
df.shape
```

```python
df.to_csv('imdb_clean_all.csv')
```

```python
# drop all the rows that year <1990
# Get names of indexes for which column startYear has value <1990
df['startYear'] = df['startYear'].astype('int')
indexNames = df[ df['startYear'] < 1990 ].index
# Delete these row indexes from dataFrame
df.drop(indexNames , inplace=True)
```

```python
df.shape
```

```python
df.to_csv('imdb_clean_1990.csv')
```

```python
#movie list from 2018 to 2020
indexNames = df[ (df['startYear'] >= 2018) & (df['startYear'] <= 2020)].index
```

```python
movie_list = {}
for i in indexNames:
    movie_name = df['primaryTitle'][i]
    movie_list[movie_name] = df['averageRating'][i]
```
